## 2. Survey design

When designing this survey I tried to keep the following survey design principles in mind:
1. Keeping the instructions as clear and concise as possible
2. Avoiding bias in the questions wording
3. Making the survey as simple as possible for the respondents to complete
4. Keep it short so as not to lose respondents attention span
5. Make the answers easy enough to analyse

I decided to use a mix of some open and some closed questions in order to gather the necessary information. I tried to limit the amount of open questions in order to reduce the complexities surrounding analysis of free-text entries. The answer choices have been mostly limited to a selection of answers. There are also some likert scale type questions that have been included which map to the main research questions designed above.

The structure of this survey is as follows:
- Demographic questions- Key demographics included in the survey are age, country of residence, industry etc
- Univariate style questions - these questions are used to gauge respondents opinions on the subject
- Likert scale questions - these map directly to the research questions which the survey aims to solve

For reference I have included a copy of the survey that respondents completed as a pdf document with the submission of this notebook.

From the outset I have also tried to be aware of how biases may arise from the study. One possible area for bias is in the sample of respondents from the overall population. This arises because the population of interest for this survey is the my network on Linkedin and because the sample has not been randomly selected to complete this survey but instead were asked to do it as a favour, there is room for bias because we only consider the answers of those who have taken the time to complete the survey. This is a major problem in all surveys and thus many survey researchers use some sort of incentive in order to entice potential respondents to complete their survey.

### Research Question 1: Are employees working remotely as efficient as they would be if they were working in an office?

```python
# Melt all columns into a single variable.
df_efficiency = df_ordinal.melt(ignore_index=False, value_vars=['I am able to stick to a work schedule.', 'I am expected to respond to emails/messages immediately.', 'I could do my job just as effectively working from home.','My company provides me with the tools and software necessary to make remote working possible.'],var_name='question', value_name='response')
# Set dtypes
df_efficiency['question'] = df_efficiency['question'].astype('category')
rating_levels = ['Strongly disagree', 'Disagree', 'Neither agree nor disagree', 'Agree', 'Strongly agree']
df_efficiency['response'] = df_efficiency['response'].astype(
    pd.CategoricalDtype(categories=rating_levels, ordered=True))
df_efficiency.head()
```

```python
# Create a normalised table (sum to 100% across rows).
table_efficiency = pd.crosstab(df_efficiency['question'], df_efficiency['response'],
                    normalize='index') * 100
# Sort by highest value.
table_efficiency.sort_values(by='Strongly agree', inplace=True)
table_efficiency.round(2)
```

```python
# Better y-axis labels.
motivations = {'I could do my job just as effectively working from home.': 'Same/more productive at home',
               'I am able to stick to a work schedule.': 'Stick to schedule',
               'I am expected to respond to emails/messages immediately.': 'Respond to emails faster',
               'My company provides me with the tools and software necessary to make remote working possible.': 'Software to work remotely'}

# Use textwrap library to break long lines automatically.
motivations = {k: '\n'.join(wrap(v, 17)) for k, v in motivations.items()}

ax = likert_plot.plot_likert_scales(table_efficiency, 'Neither agree nor disagree',
                                    qname_mapping=motivations)
ax.set_title('Questions focusing on efficency')
ax.set_ylim(1, 5)               # Manually correct y-axis.
ax.invert_yaxis()
ax.set_xlabel('Percentage')
plt.show()
```

The plot above uses the likert scale data obtained from the survey questions that map to this specific research question focusing on employee efficiency from remote working. What is interesting to note is that the majority of survey respons agreed with all questions except one that they were able to work as efficiently from home as if they had been in the office. When asked if they were expected to respond to emails faster, the answers were more spread across the board with only 48% of respondents agreeing with this statement. This serves as evidence that for the sample group of survey respondents, they are as efficient working remotely as they would be in an office environment.

### Research Question 2: Are companies able to communicate effectively with their employees?

```python
# Melt all columns into a single variable.
df_communication = df_ordinal.melt(ignore_index=False, value_vars=['My company communicates efficiently.', 'I am able to connect with my manager as well as if I was in the office.', 'I feel that I am able to easily reach my colleagues.', 'I feel that my colleagues keep me in the loop.'],var_name='question', value_name='response')
# Set dtypes.
df_communication['question'] = df_communication['question'].astype('category')
rating_levels = ['Strongly disagree', 'Disagree', 'Neither agree nor disagree', 'Agree', 'Strongly agree']
df_communication['response'] = df_communication['response'].astype(
    pd.CategoricalDtype(categories=rating_levels, ordered=True))
df_communication.head()
```

```python
# Create a normalised table (sum to 100% across rows).
table = pd.crosstab(df_communication['question'], df_communication['response'],
                    normalize='index') * 100
# Sort by highest value.
table.sort_values(by='Strongly agree', inplace=True)
table.round(2)
```

```python
# Better y-axis labels.
labels = {'My company communicates efficiently.':'Effective communication', 
          'I am able to connect with my manager as well as if I was in the office.':'Can connect with manager as well', 
          'I feel that I am able to easily reach my colleagues.':'Can easily reach colleagues', 
          'I feel that my colleagues keep me in the loop.':'Colleagues keep in touch'}

# Use textwrap library to break long lines automatically.
motivations = {k: '\n'.join(wrap(v, 17)) for k, v in motivations.items()}

ax = likert_plot.plot_likert_scales(table, 'Neither agree nor disagree',
                                    qname_mapping=labels)
ax.set_title('Questions focusing on communciation')
ax.set_ylim(1, 5)               # Manually correct y-axis.
ax.invert_yaxis()
ax.set_xlabel('Percentage')
plt.show()
```

The plot above uses the likert scale data obtained from the survey questions that map to this specific research question focusing on how effective the communication between employees and companies for those who are remote working. Once again the results of these questions are very interesting. When asked if respondents thought their company communicated effectively, 82% of respondents agreed. 79% of respondents said that they could easily reach their colleagues and 57% of them felt that their colleagues kept in touch with them. However when asked about their ability to connect with their manager while working remotely only 51% of respondents agreed. This serves as evidience that for the sample group of survey respondents, they find that communication is as effective as if they had been in the working in an office environment.

### Research Question 3: Are employees as happy working from home as they would be in the office?

```python
# Melt all columns into a single variable.
df_happiness = df_ordinal.melt(ignore_index=False, value_vars=['I get the enough support from my manager.', 'I am happy with how frequently I interact with colleagues.', 'I am happy with my work-life balance.', 'I am happy with my ability to work from home.'],var_name='question', value_name='response')
# Set dtypes.
df_happiness['question'] = df_happiness['question'].astype('category')
rating_levels = ['Strongly disagree', 'Disagree', 'Neither agree nor disagree', 'Agree', 'Strongly agree']
df_happiness['response'] = df_happiness['response'].astype(
    pd.CategoricalDtype(categories=rating_levels, ordered=True))
df_happiness.head()
```

```python
# Create a normalised table (sum to 100% across rows).
table_happiness = pd.crosstab(df_happiness['question'], df_happiness['response'],
                    normalize='index') * 100
# Sort by highest value.
table_happiness.sort_values(by='Strongly agree', inplace=True)
table_happiness.round(2)
```

```python
# Better y-axis labels.
labels = {'I get the enough support from my manager.':'Manager provides sufficent support', 
          'I am happy with how frequently I interact with colleagues.':'Enough communication from colleagues', 
          'I am happy with my work-life balance.':'Good work life-balance', 
          'I am happy with my ability to work from home.':'Happy with ability to work from home'}

# Use textwrap library to break long lines automatically.
labels = {k: '\n'.join(wrap(v, 17)) for k, v in labels.items()}

ax = likert_plot.plot_likert_scales(table_happiness, 'Neither agree nor disagree',
                                    qname_mapping=labels)
ax.set_title('Questions focusing on happiness')
ax.set_ylim(1, 5)               # Manually correct y-axis.
ax.invert_yaxis()
ax.set_xlabel('Percentage')
plt.show()
```

The plot above uses the likert scale data obtained from the survey questions that map to this specific research question focusing on how happy survey respondents are when working remotely. A massive 88% of respondents agreed with the statement that they are happy with their ability to work remotely, while 63% of respondents felt that agreed with the statement that they had a good work-life balance. 76% of respondents agreed that their manager provides them with enough support but only 51% of felt that they had enough communication from their colleagues. These results serve as evidence that working remotely does not negatively affect the happiness for the sample group of survey respondents.
